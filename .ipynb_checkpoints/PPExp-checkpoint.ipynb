```python
sets = ["Set" + str(i) for i in range(1, 6)]
labels = [str(120 + (_ - 1)*20) for _ in range(1, 6)]

for idx_s, s in enumerate(sets):
    plt.plot(getattr(NACA0021.expWTData1, s).Ave.angle,
                 -getattr(NACA0021.expWTData1, s).Ave.cl,
                 marker='o', label=labels[idx_s])

plt.title("NACA0021")
plt.xlabel("Angle $(^{\\circ})$", fontsize=15)
plt.ylabel("$c_l$ $(-)$", fontsize=15)
plt.legend(frameon=False)
```

```python
sets = ["Set" + str(i) for i in range(3, 10)]
labels = [str(120 + (_ - 3)*20) for _ in range(3, 10)]
colors = ['k', 'g', 'r', 'b', 'orange']

plt.figure()

for idx_geom, geom in enumerate(["DU96W180", "DU00W212", "DU91W250", "DU97W300", "DU00W401"]):
    obj = globals()[geom]

    for idx_s, s in enumerate(sets):
        if idx_s == len(sets) - 1:            
            plt.plot(getattr(obj.expWTData1, s).Ave.angle,
                         -getattr(obj.expWTData1, s).Ave.cl,
                         marker='o', color=colors[idx_geom], alpha=(idx_s + 1)/len(sets),
                         label=geom)
        else:
            plt.plot(getattr(obj.expWTData1, s).Ave.angle,
                         -getattr(obj.expWTData1, s).Ave.cl,
                         marker='o', color=colors[idx_geom], alpha=(idx_s + 1)/len(sets))            

plt.xlabel("Angle $(^{\\circ})$", fontsize=15)
plt.ylabel("$c_l$ $(-)$", fontsize=15)
plt.legend(frameon=False)
plt.title("Airfoil-wise comparative")
plt.show()
```

```python
import pandas as pd

aoa = np.arange(-5, 21)

data_exp = {}
for dataset_str in ["DU96W180", "DU00W212", "DU91W250", "DU97W300", "DU00W401"]:
    dataset = globals()[dataset_str]
    sheet_name = dataset_str + "_exp"
    data_exp["AoA"] = aoa
    for i in range(3, 10):
        data_exp["Re" + str(80 + (i - 1)*20)] = -getattr(dataset.expWTData1, "Set" + str(i)).Ave.cl
    # Create a DataFrame
    df = pd.DataFrame(data_exp)
    if os.path.exists("output_exp.xlsx"):
        # Append mode (file already exists)
        with pd.ExcelWriter("output_exp.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
            df.to_excel(writer, sheet_name=sheet_name, index=False)
    else:
        # Create new file
        with pd.ExcelWriter("output_exp.xlsx", engine="openpyxl") as writer:
            df.to_excel(writer, sheet_name=sheet_name, index=False)
```
